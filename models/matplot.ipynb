```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mpl.style.use('bmh')
```

```python
# This line helps because the background color of my vsc is black
# Otherwise I cannot seer the labels in the plots.
plt.rcParams['figure.facecolor'] = 'white'
```

```python
# Let's plot a simple line between two pairs of points.
# Matplot has an object and state interface. We will use the object one
fig, ax = plt.subplots()
ax.plot([1, 3], [2, 4])
# Basically, we first create a figure, which can then have several axes.
# Also, we can change proprties of the axes
ax.set_xlabel('This is the X Axis label')
ax.set_ylabel('This is the Y Axis label')
ax.set_title('This is the title of the graph')
ax.legend
```

```python
# Another way to create a figure is via the figure method
%matplotlib inline
fig, ax = plt.figure(figsize = (5, 5))
ax.plot([5,48,6,54,5,6,8,6,2,99,6,5])
```
